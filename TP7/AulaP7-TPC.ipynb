Cria uma aplicação Python que permita ao utilizador usar todas as funcionalidades pedidas nesta ficha.


```python
def criatabMeteo(n):
    i = 0
    tabMeteo = []
    while i < n:
        ano = int(input("Insira o ano:")) 
        mes = int(input("Insira o mes:"))
        dia = int(input("Insira o dia:"))
        data = [ano,mes,dia]
        min = float(input("Insira a temperatura miníma:"))
        max = float(input("Insira a temperatura máxima:"))
        prec = float(input("Insira a precipitação:"))
        dia = (data,min,max,prec)
        tabMeteo.append(dia)
        i = i + 1
        print(f"Dia inserido com sucesso{dia}")
    return tabMeteo

def medias(tabMeteo):
    res = []
    for data,tempMin,tempMax,_ in tabMeteo:
        media = (tempMin + tempMax) / 2
        res.append((data,media))
    return res

def guardaTabMeteo(t, fnome):
    f = open(fnome, "w")
    for data, tempMin, tempMax, prec in t:
        ano, mes, dia = data
        f.write(f"{ano}-{mes}-{dia}|{tempMin}|{tempMax}|{prec}\n")
    return

def carregaTabMeteo(fnome):
    res = []
    f = open(fnome,"r")
    for line in f:
        line = line.strip()
        campos = line.split("|")
        data, min, max, prec = campos
        ano, mes, dia = data.split("-")
        tuplo = ((int(ano), int(mes), int(dia)), float(min), float(max), float(prec))
        res.append(tuplo)
    return res

def minMin(tabMeteo):
    minima = tabMeteo[0][1]
    for _, min, *_ in tabMeteo[1:]:
        if min < minima:
            minima = min
    return minima

def amplTerm(tabMeteo):
    res = []
    for data, min, max, _ in tabMeteo:
        amp = max - min
        tuplo = (data,amp)
        res.append(tuplo)
    return res 

def maxChuva(tabMeteo):
    max_prec = tabMeteo[0][3]
    for dia in tabMeteo[1:]:
        if dia[3] > max_prec:
            max_prec = dia[3]
            max_data = dia[0] 
    return (max_data, max_prec)

def diasChuvosos(tabMeteo, p):
    res = []
    for data, *_, prec in tabMeteo:
        if prec > p:
            res.append((data,prec))
    return res

def maxPeriodoCalor(tabMeteo, p):
    consecutivos = 0
    consecutivos_global = 0
    for dia in tabMeteo:
        if dia[3] < p:
            consecutivos = consecutivos + 1
        else:
            if consecutivos > consecutivos_global:
                consecutivos_global = consecutivos
            consecutivos = 0
    
    if consecutivos > consecutivos_global:
        consecutivos_global = consecutivos
    return consecutivos_global

def menu():
    print("""
    ------- MENU -------
    1: Criar tabela meterológica;
    2: Inserir automaticamente uma tabela meteorológica;
    3: Calcular a temperatura média de cada dia;
    4: Guardar a tabela meteorológica num ficheiro de texto;
    5: Carregar a tabela meteorológica de um ficheiro de texto;
    6: Calcular a temperatura mínima mais baixa registada na tabela;
    7: Calcular a amplitude térmica de cada dia;
    8: Calcular o dia em que a precipitação registada teve o seu valor máximo;
    9: Dias em que a precipitação foi superior a um certo valor;
    10: Maior número consecutivo de dias com precipitação abaixo de um certo limite;
    0: Sair da aplicação.
    """)

tabMeteo = []
cond = True
print("Certifique-se que existem dados na tabela meteoróligica com as opções 1 ou 2.")
menu()
while cond:
    opcao = int(input("Escolha uma opção: "))
    if opcao == 1:
        n = int(input("Insira o número de dias que vai inserir na tabela:"))
        tabMeteo = criatabMeteo(n)    
    elif opcao == 2:
        tabMeteo = [((2022,1,20), 2, 16, 0),((2022,1,21), 1, 13, 0.2), ((2022,1,22), 7, 17, 0.01)]
        print(f"Tabela meteorológica inserida com sucesso: {tabMeteo}")
    elif opcao == 3:
        print(medias(tabMeteo))
    elif opcao == 4:
        guardaTabMeteo(tabMeteo, "meteorologiaTPC.txt")
        print("Tabela guardada com sucesso")
    elif opcao == 5:
        tabMeteo = carregaTabMeteo("meteorologiaTPC.txt")
        print(f"Tabela carregada com sucesso{tabMeteo}")
    elif opcao == 6:
        print(minMin(tabMeteo))
    elif opcao == 7:
        print(amplTerm(tabMeteo))
    elif opcao == 8:
        print(maxChuva(tabMeteo))
    elif opcao == 9:    
        p = float(input("Insira o valor de precipitação para o qual quer ver os dias em que a mesma foi superior:"))
        print(diasChuvosos(tabMeteo, p))
    elif opcao == 10:
        p = int(input("Insira o valor de precipitação para o qual quer ver os dias consecutivos em que a mesma foi inferior:"))
        print(maxPeriodoCalor(tabMeteo, p))
    elif opcao == 0:
        cond = False
        print("Saindo do programa...")
    else:
        print("Opção inválida! Tente novamente.")
```
