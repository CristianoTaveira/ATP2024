```python
import random

def decisao():
    while True:
        modalidade = int(input("Escolha a modalidade que quer jogar (1/2): "))
        if modalidade in [1, 2]:
            return modalidade
        else:
            print("Opção inválida. Escolha 1 ou 2.")

def menu():
    print("""--------------------- MENU --------------------
Modalidade 1: Jogador joga em primeiro lugar
Modalidade 2: Computador joga em primeiro lugar""")
    return decisao()

def loop():
    opcao = input("Deseja sair(s/n)?: ")
    while opcao not in ("s", "n"):
        opcao = input("Deseja sair(s/n)?: ")
    if opcao == "s":
        print("Obrigado, até à próxima")
    elif opcao == "n":
        main()

def jogada_computador(f):
    if f == 1:
        print("Ficou 1 fósforo. Sou obrigado a tirar o último fósforo.")
        return 0  
    n = (f - 1) % 5
    if n == 0:
        n = random.randint(1, 4)
    if f - n == 1:
        print(f"Tirei {n}, ficou {f - n} fósforo")
    else:
        print(f"Tirei {n}, ficaram {f - n} fósforos")
    return f - n


def jogada_jogador(f):
    escolha = int(input(f"Fósforos restantes:{f}, quantos queres tirar (1, 2, 3 ou 4): "))
    while escolha not in [1, 2, 3, 4] or escolha > f:
        if f == 1:
            escolha = int(input(f"Escolha inválida. Existe {f} fósforo, só pode tirar 1"))
        else:
            escolha = int(input(f"Escolha inválida. Existem {f} fósforos, quantos queres tirar (1, 2, 3 ou 4): "))
    print(f"Tirou {escolha}, fósforos restantes:{f - escolha}")
    return f - escolha

    

def modalidade1():
    f = 21
    while f > 0:  
        f = jogada_jogador(f)
        if f == 0:
            print("Ganhei!")
            return
        f = jogada_computador(f)

def modalidade2():
    f = 21
    f = jogada_computador(f)  
    while f > 0:
        f = jogada_jogador(f)
        if f == 0:
            print("Ganhei!")
            return
        f = jogada_computador(f)
        if f == 0:
            print("Parabéns, ganhaste!")
            return

def main():
    modalidade = menu()
    print("Existem 21 fósforos, apenas se pode tirar 1, 2, 3 ou 4 em cada jogada, quem ficar com o último fósforo perde, boa sorte!")
    if modalidade == 1:
        modalidade1()
    elif modalidade == 2:
        modalidade2()
    loop()

main()
```
