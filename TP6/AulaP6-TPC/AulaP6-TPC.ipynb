### TPC6: Aplicação para gestão de alunos

Considere que o modelo do aluno e da turma têm a seguinte estrutura:

`aluno = (nome, id, [notaTPC, notaProj, notaTeste])`

`turma = [aluno]`

* Cria uma aplicação que coloca no monitor o seguinte menu de operações:
    - 1: Criar uma turma;
    - 2: Inserir um aluno na turma;
    - 3: Listar a turma;
    - 4: Consultar um aluno por id;
    - 8: Guardar a turma em ficheiro;
    - 9: Carregar uma turma dum ficheiro;
    - 0: Sair da aplicação
* No fim de executar a operação selecionada, a aplicação deverá colocar novamente o menu e pedir ao utilizador a opção para continuar;
* Utiliza a tua aplicação para criar uma turma com 5 alunos.

```python
def inserirAluno(turma):
    nome = input("Insira o nome do aluno: ")
    id = input("Insira o id do aluno: ")
    tpc = int(input("Insira a nota do TPC: "))
    proj = int(input("Insira a nota do projeto: "))
    teste = int(input("Insira a nota do teste: "))
    notas = [tpc, proj, teste]
    aluno = (nome, id, notas)
    turma.append(aluno)
    return turma

def listarTurma(turma):
    print("(nome, id, [notaTPC, notaProjeto, notaTeste])")
    for aluno in turma:
        print(aluno)

def existeId(turma, id):
    for aluno in turma:
        if id == aluno[1]:
            return True
    return False

def consultarById(turma, id):
    for aluno in turma:
        if id == aluno[1]:
            print(aluno)
            return aluno

def guardar(turma):
    with open("turma.txt", "w") as f:
        for aluno in turma:
            nome, id, notas = aluno
            tpc, proj, teste = notas
            f.write(f"{nome}#{id}#{tpc}#{proj}#{teste}\n")
    print("Turma guardada com sucesso no ficheiro turma.txt")

def carregar(fnome):
    turma = []
    with open(fnome, "r") as file:
        for line in file:
            line = line.strip()
            campos = line.split("#")
            nome = campos[0]
            id = campos[1]
            notas = [int(campos[2]), int(campos[3]), int(campos[4])]
            aluno = (nome, id, notas)
            turma.append(aluno)
    print("Turma carregada com sucesso do ficheiro turma.txt")
    return turma

def menu():
    print("""
    ------- MENU -------
    1: Criar uma turma;
    2: Inserir um aluno na turma;
    3: Listar a turma;
    4: Consultar um aluno por id;
    5: Guardar a turma em ficheiro;
    6: Carregar uma turma dum ficheiro;
    0: Sair da aplicação.
    """)

cond = True
while cond:
    menu()
    opcao = int(input("Escolha uma opção: "))
    if opcao == 1:
        turma = []
        print("Turma criada com sucesso")
    elif opcao == 2:
        inserirAluno(turma)
        print("Aluno inserido com sucesso")
    elif opcao == 3:
        listarTurma(turma)
    elif opcao == 4:
        id = input("Insira o id do aluno que deseja procurar (exemplo: a100000): ")
        if existeId(turma, id):
            consultarById(turma, id)
        else:
            print("Não existe um aluno com esse id")
    elif opcao == 5:
        guardar(turma)
    elif opcao == 6:
        turma = carregar("turma.txt")
    elif opcao == 0:
        cond = False
        print("Saindo do programa...")
    else:
        print("Opção inválida! Tente novamente.")
```
